```python
%matplotlib inline
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('image.png')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray, cmap='gray')
```

# Face Detection with Haar Cascade

```python
face_detector = cv2.CascadeClassifier('haarcascade_frontalface_alt.xml')
```

```python
faces = face_detector.detectMultiScale(img_gray)
```

```python
faces
```

```python
vis_img = img.copy()
for x, y, w, h in faces:
    cv2.rectangle(vis_img, (x, y), (x + w, y + h), (0, 255, 0), 2)
```

```python
plt.imshow(vis_img[:,:,::-1])
```

# Eyes Detection with Haar Cascade

```python
eye_detector = cv2.CascadeClassifier('haarcascade_eye_tree_eyeglasses.xml')
eyes = eye_detector.detectMultiScale(img_gray)
vis_img2 = img.copy()
for x, y, w, h in eyes:
    cv2.rectangle(vis_img2, (x, y), (x + w, y + h), (0, 255, 0), 2)
plt.imshow(vis_img2[:,:,::-1])
```

```python
vis_img3 = img.copy()
for x, y, w, h in eyes:
    cv2.rectangle(vis_img3, (x, y), (x + w, y + h), (255, 0, 0), 2)
for x, y, w, h in faces:
    cv2.rectangle(vis_img3, (x, y), (x + w, y + h), (0, 255, 0), 2)
plt.imshow(vis_img3[:,:,::-1])
```
